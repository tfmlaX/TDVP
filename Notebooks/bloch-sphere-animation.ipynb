```python
import csv

pop2 = []
c12R = []
c12I = []

with open('test.csv') as csvDataFile:
    csvReader = csv.reader(csvDataFile)
    for row in csvReader:
        pop2.append(row[0])
        c12R.append(row[1])
        c12I.append(row[2])

pop2 = list(map(float, pop2))
c12R = list(map(float, c12R))
c12I = list(map(float, c12I))
```

```python
u = 2*np.array(c12R);
v = -2*np.array(c12I);
w = 1 - 2*np.array(pop2);
```

```python
fig = plt.figure(figsize=(8,8))
axes = Axes3D(fig, azim=-40,elev=30)
sphere = Bloch(axes=axes)
sphere.add_points([u, v, w], meth='l')
sphere.make_sphere()
sphere.show()
```

```python
b = Bloch()

b.view = [-40,30]

for i in range(len(u)):
    b.clear()
    b.add_points([u[:i+1],v[:i+1],w[:i+1]])
    b.save(dirc='temp') #saving images to temp directory in current working directory
```
