## Time Series

```julia
times = data["times"];

# Sites populations
site_pop = Array{Float64}(undef, N, length(times))

for n=1:N-1
    site_pop[n,:] = data["n$(n)"]
end

# Coherences
coherence12 = data["coherence12"]
if N>2
    coherence23 = data["coherence23"]
    coherence13 = data["coherence13"]
end

# System's Energy Eigen State Population
if N==1 # SBM
    eigen_pop2 = 0.5 .+ real.(coherence12)
    eigen_pop1 = 1 .- eigen_pop2

elseif N==2
    Ea = 0; Eb = 0.;
    H = [Ea ω0; ω0 Eb] # System Hamiltonian
    F = eigen(H) 
    eival = real(F.values)
    I = sortperm(eival) # give the indices of the eigenvalues sorted 
    eival = eival[I]
    eivec = F.vectors[:,I]
    
    eigen_pop1 = abs(eivec[1,1])^2*site_pop[1,:] + abs(eivec[2,1])^2*site_pop[2,:]  + 2*real(eivec[1,1]*conj(eivec[2,1])*coherence12)
    eigen_pop2 = 1 .- eigen_pop1
    #eigen_pop2 = 0.5 .+ real.(coherence12)
    #eigen_pop1 = 1 .- eigen_pop2
elseif N==3
# Eigen-states populations 3 sites
    Ea = 0; Eb = -2*α; Ec = -1;
    H = [Ea ω0 0; ω0 Eb ω0; 0 ω0 Ec] # System Hamiltonian
    F = eigen(H) 
    eival = real(F.values)
    I = sortperm(eival) # give the indices of the eigenvalues sorted 
    eival = eival[I]
    eivec = F.vectors[:,I]
    
    eigen_pop1 = abs(eivec[1,1])^2*site_pop[1,:] + abs(eivec[2,1])^2*site_pop[2,:] + abs(eivec[3,1])^2*site_pop[3,:] + 2*real(eivec[1,1]*conj(eivec[2,1])*coherence12 + eivec[1,1]*conj(eivec[3,1])*coherence13 + eivec[2,1]*conj(eivec[3,1])*coherence23)
    eigen_pop2 = abs(eivec[1,2])^2*site_pop[1,:] + abs(eivec[2,2])^2*site_pop[2,:] + abs(eivec[3,2])^2*site_pop[3,:] + 2*real(eivec[1,2]*conj(eivec[2,2])*coherence12 + eivec[1,2]*conj(eivec[3,2])*coherence13 + eivec[2,2]*conj(eivec[3,2])*coherence23)
    #eigen_pop2 = 0.5.*(site_pop[1,:] + site_pop[3,:]) .- real.(data["coherence13"])
    #eigen_pop1 = 0.25.*(data["n1"] + 2.0.*data["n2"] + data["n3"]) + 0.5.*(real.(data["coherence13"]) - sqrt(2).*(real.(data["coherence12"]) + real.(data["coherence23"])))
    eigen_pop3 = 1 .- eigen_pop1 .- eigen_pop2
    
elseif N>3+1
    # Construction of the Hamiltonian matrix
    Esites = [1 0.5 0.75 0]
    Etunnel = [ω0]
    if length(Etunnel)==1
        H = Etunnel[1].*ones(N,N)
        for i=1:N
            H[i,i] = Esites[i]
        end
    else
        for i=1:N
            H[i,i] = Esites[i]
            for j=i+1:N
                H[i,j] = H[j,i] = Etunnel[i,j]
            end
        end
    end
    
    # Diagonalization
    F = eigen(H) 
    eival = real(F.values)
    Ind = sortperm(eival) # give the indices of the eigenvalues sorted 
    eival = eival[Ind]
    eivec = F.vectors[:,Ind]
    
    # Construction of eigen populations
    eigen_pop = Array{Float64}(undef, N, length(times))
    # First we construct ρ in the site basis
    ρ = Array{ComplexF64}(undef, N, N, length(times))
    coherence = Array{ComplexF64}(undef, N-2, N-1, length(times))
    coherence[1,2,:] = coherence12 
    coherence[1,3,:] = coherence13
    coherence[1,4,:] = coherence14
    coherence[2,3,:] = coherence23
    coherence[2,4,:] = cpherence24
    coherence[3,4,:] = coherence34
    for i=1:N
    ρ[i,i,:] = site_pop[i,:]
       for j=i+1:N
            ρ[i,j,:] .= coherence[i,j,:]
            ρ[j,i,:] .= conj(ρ[i,j,:])
       end
    end
    # The eigen populations are the matrix element <eig-vec|ρ|eig-vec>
    for k=1:N
        for t=1:length(times)
            eigen_pop[k,t] = real(dot(eivec[:,k],ρ[:,:,t]*eivec[:,k]))
        end
    end    
    
end
# Chain modes occupation
bathOcc = data["bathOcc"];
```

### Bath Displacement from System's observables

```julia
ksample = [-1:0.001:1...];
```

```julia
gk = [sqrt(2*α*abs(k)) for k in ksample];
```

```julia
ak = zeros(ComplexF64, length(ksample), 2*length(times));
```

```julia
for p=1:length(ksample)
    for j=1:2*length(times)
        S = 0
        dt = times[2]-times[1]
        for n=1:N
            LaplaceOfPop = 0
            for l=1:j
                if l <= length(times) && n==2
                    LaplaceOfPop += exp(im*abs(ksample[p])*c_phonon*times[l])*dt
                elseif l>length(times)
                    LaplaceOfPop += exp(im*abs(ksample[p])*c_phonon*(times[end]+times[l-length(times)]+dt))*site_pop[n,l-length(times)]*dt
                end
            end
            
            S += exp(-im*ksample[p]*R*(n-1))*LaplaceOfPop
        end
        
        ak[p,j] = j <= length(times) ? -im*gk[p]*exp(-im*abs(ksample[p])*c_phonon*times[j])*S : -im*gk[p]*exp(-im*abs(ksample[p])*c_phonon*(times[j-length(times)]+times[end]+dt))*S
    end
end
```

```julia
space = [-60:1:100...];
```

```julia
displacement = zeros(length(space), 2*length(times));
```

```julia
for x=1:length(space)
    for t=1:2*length(times)
        for k=1:length(ksample)
            if ksample[k]==0
                continue
            else
                displacement[x,t] += 2*real(ak[k,t]*exp(im*ksample[k]*space[x]))/sqrt(abs(ksample[k])*c_phonon*length(space))
            end
        end
    end
end
```

```julia
max_val = maximum(abs.(displacement));
```

```julia
heatmap(space, doubletime, displacement', color=:balance, clims=(-max_val,max_val))
xlabel!(L"x")
ylabel!(L"\omega_c t")
```

```julia
function wavepacket(modesOcc, ksamples, x, t_ind, c=1)
    S = 0
    for i=1:length(ksample)
        S += modesOcc[t_ind][i]*exp(im*ksample[i]*(x-times[t_ind]*c))
    end
    return S
end
```
